```python
import torch
from torch import nn
import torch.optim as optim
from sklearn.metrics import accuracy_score
```

```python
m = nn.Linear(5, 3)
input = torch.randn(4, 5)
output = m(input)

print('Input:', input, f'shape {input.shape}', sep='\n')
print('\nOutput:', output, f'shape {output.shape}', sep='\n')
```

```python
print("Weights:\n", m.weight)
print("Bias:\n", m.bias)
```

```python
t = torch.randn(4)
print('Input: ', t)
print('Applying sigmoid: ', torch.sigmoid(t))
```

```python
import matplotlib.pyplot as plt

randn_data = torch.randn(10000)  # Нормальний розподіл
rand_data = torch.rand(10000)    # Рівномірний розподіл

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(randn_data.numpy(), bins=50, color='blue', alpha=0.7)
plt.title("torch.randn() - Нормальний розподіл")

plt.subplot(1, 2, 2)
plt.hist(rand_data.numpy(), bins=50, color='green', alpha=0.7)
plt.title("torch.rand() - Рівномірний розподіл")

plt.show()
```
